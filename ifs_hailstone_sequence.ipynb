# Plotting the Hailstone Sequence
Hill's OpenGL book 2nd, p. 71.


```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
def f(x):
    if x % 2 == 0:  # even
        return x // 2
    else:
        return 3 * x + 1  # odd 
```

```python
x = np.random.randint(0, 4000_000_000, dtype='int64')
s = [x]
while True:
    x = f(x) 
    s.append(x)
    if x == 1: break 
print(len(s))
```

```python
print(s[0], max(s), min(s), len(s))
plt.plot(np.log10(s));
```

```python
print(s[0], max(s), min(s), len(s))
plt.plot(s);
plt.ylim([0, 100]);
```
