```python
import cv2 
import skimage
import imageio
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
filename = 'pixels/image-15.png'
im = imageio.imread(filename)
im = skimage.data.coffee()
plt.imshow(im);
# plt.show()
```

```python
im1 = im[:, 350]  # red channel at x=350
im1.shape
color = ['red', 'green', 'blue']
for i in range(3):
    plt.plot(im1[:,i], label=f'c={i}', color=color[i])
plt.legend();
```

```python
r0, r1 = 0, 10  #300, 320
for i in range(3):
    plt.plot(im1[r0:r1,i], label=f'c={i}', color=color[i])
plt.legend();
```

```python
plt.stem(im1[r0:r1,2], label=f'blue')
plt.legend();
```

- INTER_NEAREST - a nearest-neighbor interpolation
- INTER_LINEAR - a bilinear interpolation (used by default)
- INTER_AREA - resampling using pixel area relation. It may be a preferred method for image decimation, as it gives moire-free results. But when the image is zoomed, it is similar to the INTER_NEAREST method.
- INTER_CUBIC - a bicubic interpolation over 4x4 pixel neighborhood
- INTER_LANCZOS4 - a Lanczos interpolation over 8x8 pixel neighborhood


```python
scale_y = 5
interpolation = cv2.INTER_LINEAR
interpolation = cv2.INTER_NEAREST
im_scaled = cv2.resize(im, dsize=(0,0), fx=1, fy=scale_y, interpolation=interpolation)
plt.imshow(im_scaled)
```

```python
im1_scaled = im_scaled[:, 350]  # red channel at x=350
color = ['red', 'green', 'blue']
sr0, sr1 = r0 * scale_y, r1 * scale_y 
for i in range(3):
    plt.plot(im1_scaled[sr0:sr1, i], label=f'c={i}', color=color[i])
plt.legend();
```

```python
plt.stem(im1_scaled[sr0:sr1,2], label=f'scaled')

rid = np.arange(len(im1[r0:r1]))
print(rid)
plt.stem(2+rid*scale_y,im1[r0:r1,2], 'red', markerfmt='ro', label='org')
plt.ylim([34, 44])
plt.legend();
```
