# IFS Orbits of sin/cos
1. the orbit of 1 for the function $\sin(\cdot)$
2. the orbit of 0 for the function $\cos(\cdot)$
3. the orbit of 1 for the function $\cos(\cdot)$

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)
```

```python
x = 1  # initial
orbit = [x]
for _ in range(1000):
    x = np.sin(x)
    orbit.append(x)
```

```python
plt.plot(orbit)
plt.title("the orbit of 1 for sin()");
```

```python
x = 1
orbit1 = [x]
for _ in range(1000):
    x = np.cos(x)
    orbit1.append(x)
```

```python
plt.plot(orbit1[:30])
plt.title(f"the orbit of {1} for cos()");
```

```python
x = 0
orbit0 = [x]
for _ in range(1000):
    x = np.cos(x)
    orbit0.append(x)
```

```python
plt.plot(orbit0[:30])
plt.title(f"the orbit of {0} for cos()");
```

```python
plt.plot(orbit1[:20], label="orbit of 1")
plt.plot(orbit0[:20], label="orbit of 0")
plt.legend();
```
